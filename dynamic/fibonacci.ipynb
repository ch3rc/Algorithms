```
import math
import sys
import time
import matplotlib.pyplot as plt

x = []
x2 = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []


# recursive approach O(2^n) runtime
def RECURSIVE_FIB(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return RECURSIVE_FIB(n - 2) + RECURSIVE_FIB(n - 1)

# top down optimize approach ()
def TD_REC_FIB(n, r):
  if (r[n] == None):
    if n < 2:
      r[n] = n
    else:
      r[n] = TD_REC_FIB(n - 1, r) + TD_REC_FIB(n-2, r)
  return r[n]

def TD_FIB(n):
  r = [None] * (n + 2)
  return TD_REC_FIB(n, r)

def BU_FIB(n):
  r = [None] * (n + 2)
  r[0] = 0
  r[1] = 1
  for i in range(2, n+2):
    r[i] = r[i - 1] + r[i - 2]
  return r[n]



for i in range(1, 40):
  x.append(i)

  s1 = time.time()
  RECURSIVE_FIB(i)
  e1 = time.time()
  y1.append(e1 - s1)
  
  s2 = time.time()
  TD_FIB(i)
  e2 = time.time()
  y2.append(e2 - s2)

  s3 = time.time()
  BU_FIB(i)
  e3 = time.time()
  y3.append(e3 - s3)


for i in range(1, 700):
  x2.append(i)
  
  s4 = time.time()
  TD_FIB(i)
  e4 = time.time()
  y4.append(e4 - s4)

  s5 = time.time()
  BU_FIB(i)
  e5 = time.time()
  y5.append(e5 - s5)



plt.plot(x, y1, label="Recursive approach", color="green")
plt.plot(x, y2, label="Top-down approach", color="blue")
plt.plot(x, y3, label="Bottom-up approach", color="red")
plt.xlabel('Iterations')
plt.ylabel('Run time')
plt.title('Dynamic programming - Fibonacci Sequence')
plt.legend()
plt.show()


plt.plot(x2, y4, label="Top-down approach", color="blue")
plt.plot(x2, y5, label="Bottom-up approach", color="red")
plt.xlabel('Iterations')
plt.ylabel('Run time')
plt.title('Dynamic programming - Fibonacci Sequence - TD and BU only')
plt.legend()
plt.show()
```
